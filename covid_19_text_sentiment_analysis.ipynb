```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus=[]

for i in range(0, 79):
  title = re.sub('(@[A-Za-z0-9]+)|(#[A-Za-z0-9]+)|([^0-9A-Za-z t])|(w+://S+)', ' ', df['title'][i])
  title = title.split("https")[0]
  title = title.lower()
  title = title.split()
  title = [ps.stem(word) for word in title if not word in set(all_stopwords)]
  title = ' '.join(title)
  corpus.append(title)
```

```python
df['Text'] = corpus
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
df['scores'] = df['Text'].apply(lambda Text: sid.polarity_scores(Text))
df.head()
```

```python
 df[['title','scores']]
```

```python
df = pd.concat([df, df['scores'].apply(pd.Series)], axis=1)
```

```python
df[['title','neg', 'neu', 'pos', 'compound']]
```
